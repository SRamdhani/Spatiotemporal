```python
from sklearn.decomposition import PCA
from st.sim import Simulation
from st.model import Model
from st.plots import Plots
from st.util import Utility
import numpy as np
```

```python
size = 200 # Total number of sampled points.
comp = 50  # Total number of components 
```

#### 1. Create SpatioTemporal by creating spatial data then layering on a Time Series AR-1 Process

```python
# Spatial data
sim_data, sample_points = Simulation.create_dataset(sim_num=1, size=size, seeds=234)

# Time Series
ar_fl_time_series = Simulation.ar1(it_T=100)
```

```python
# Plot Time Series
Plots.basic_time_plot(ar_fl_time_series)
```

```python
# Create SpatioTemporal Data: True Data and Data with Noise
st_data_true, st_data_noise = Utility.add_time_to_spatial(ar_fl_time_series, 
                                                  np.expand_dims(sim_data, 1))
```

```python
print(st_data_true.shape)
print(st_data_noise.shape)
```

```python
# The real data set means that which would be observed in real life, hence it 
# would have noisy measurements.
sim_real = st_data_noise
sim_real.shape
```

#### 2. Center and Get EOF components and Check them.

```python
# 1. Spatially Centered

sim_centered = sim_real - sim_real.mean(1).reshape(-1, 1)
sim_centered.shape
```

```python
# 2. EOF

pca = PCA(comp)
pca.fit(sim_centered)
principalComponents = pca.transform(sim_centered)
```

```python
print(principalComponents.shape) # time (rows) x comp (cols)
print(pca.components_.shape)     # comp (rows) x location (cols)
```

```python
# Recreate the centered data using 5 components. 
Utility.check_pca(principalComponents, pca, 5)
```

```python
# Compare this to the recreated version above.
sim_centered
```

#### 3. Create Neural Network Model, Train and Recreate the Field

```python
# Instantiate & Build Model
model = Model(comp=comp)
model.build_model()
```

```python
# Set up data generator
train_dataset, principalComponentsTensor = Utility.create_training_data_tf(principalComponents, pca, sample_points)

# Train Model
model.training(train_dataset,
               principalComponentsTensor,
               lr=1e-3, epoch_num=120) 
```

```python
true, recon = Utility.get_true_recon_field(principalComponents, pca,
                                           model.predict(sample_points)) 
```

#### 4. Check time series prediction and spatial prediction at a specific location.

```python
# Plotting on the first 5 plots. Max number of plots would be size variable.
Plots.true_vs_recon_time_series(true, recon, sample_points,
                                total_plots = 5) 
```

```python
# Plotting the spatial heatmap for the true and reconstructed field from the neural network. 
# Plotting only the first 5 plots. 
# The loc variable is one of the sampled points marked with a white X to compare the true value
# and the reconstructed or predicted value.
Plots.spatial_heatmap(true, recon, sample_points,
                      loc = 50, total_plots = 5)
```
